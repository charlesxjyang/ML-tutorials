## Abstraction
One of the most powerful concepts in Computer Science is **Abstraction**. Simply put, Abstraction is the idea that we can ignore how methods are implemented - all we need to understand are the inputs and outputs. Abstraction is a major contributing factor to the proliferation of open source packages and software. Rather than focusing on how the underlying code works, we simply **trust** that it will work. As a result, incredibly complex software can be used by anyone who understands the inputs and outputs. It is also partially responsible for the highly collaborative and open-source nature of CS. Simply by sharing a few kb of text files can save others development time and allow anyone anywhere to reproduce and build upon your work.

Now it might seem like a slightly scary idea to trust code written by people you've never met. Even more frightening, we don't even know how it works! Of course, there are caveats. We should always ensure that the source code is reliable(number of users/developers involved is generally a good metric) - definitely don't go around the interwebs downloading every snippet of code you find. But generally, abstraction is a powerful tool that can save us immense amounts of time - you just have to be willing to be a little more uncertain and aware.

This notebook will draw on abstraction very heavily, probably slightly more than is healthy, in the hope that the resulting material will be simple enough for people of all levels to understand quickly. For the most part, we will trust abstraction to do the heavy lifting for us and that the various packages we use do what they promised us.

```python
def adder(x):
    return x+7
def multiplier(x):
    return 7*x
```

```python
#Treating functions as inputs
def composition_function(f,x):
    return f(f(f(x)))
composition_function(multiplier,3)
```

```python
#Fib-o-nacci Fun
#This function returns the nth element in the Fibonacci Sequence
def fib(n):
    a,b = 1,1
    for i in range(n-1):
        a,b = b,a+b
    return a
#Higher Order Functions - Functions inside of functions
def HOF_fib(n):
    #We can define a helper function inside of another function
    def fib_helper(a,b,n):
        if n==0:
            return a
        else:
            return fib_helper(b,a+b,n-1) #this method is actually called tail recursion
    return fib_helper(0,1,n)

#Another example of passing functions in as inputs
def fnx_comparer(x,f1,f2):
    '''Compares f1(x) to f2(x) and determines if they are equivalent'''
    print('f1 of ' + str(x) + ": " + str(f1(x)))
    print('f2 of ' + str(x) + ": " + str(f2(x)))
    if f1(x)==f2(x):
        print('f1 of ' + str(x) + ' and f2 of ' + str(x) + ' are equal')
    else:
        print('f1 of ' + str(x) + ' and f2 of ' + str(x) + ' are NOT equal')
fnx_comparer(10,fib,HOF_fib)
```

## Loops


```python
for i in range(10):
    print(i)
```

```python
i = 0
while i < 10:
    print(i)
    i = i+1
```

```python
#Zip takes in n number of iterables(e.g. lists,ranges) and 'zips' them together into one big iterable
for i,j in zip(range(10),range(10,20)):
    print("This is: " + str(i))
    print("This is: " + str(j))
    print("\n")
```

```python
#One common usage of while loops is an external counter
#Can you figure out what the cell below is doing?
counter = 0
run_sum = 0
while run_sum<1000:
    run_sum = run_sum + counter
    counter = counter+1
print(run_sum)
print(counter)
```

You should be aware that using loops is almost always the **most inefficient** way to accomplish a task. It is always better to "vectorize" your code instead of simply looping over it. ## Scikit Learn
Scikit Learn is one of the most popular high-level API's for machine learning and data analysis in Python. Designed to be as succinct and clean as possible, the entire library revolves around the model object. In this section, I'm simply going to walk through 2 models, one is logistic regression(essentially linear regression for classification problems) and Random Forests for predicting diagnoses based on the features of the tumor. Note that you don't actually have to understand how either of these work. What I want to demonstrate is the versatilty of the Model API, which leverages abstraction, to allow anyone with basic understanding of programming utilize incredibly powerful and discerning models.

The confusion matrix plotter was adapted from [sklearn example](http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html#sphx-glr-auto-examples-model-selection-plot-confusion-matrix-py)

```python
#Adopted from sklearn confusion_matrix example
def plot_confusion_matrix(y_test,y_pred, classes):
    import itertools
    from sklearn.metrics import confusion_matrix
    #sets the color scale
    cmap=plt.cm.Blues
    #runs the confusion matrix on the predicted labels vs actual labels
    cm = confusion_matrix(y_test,y_pred)
    cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    print("Normalized confusion matrix")
 
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title('Confusion matrix')
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    #setting the tick labels
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)
    #determining color/alignment of ticklabels
    fmt = '.2f'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
#converts M,B to 1,0 respectively
y = np.array(df['diagnosis']=='M').astype(int)

#split our dataset up into a training and test set
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,1:],df['diagnosis'])

#instantiating the model object
model = LogisticRegression()
#fitting/training model on train data
model.fit(X_train,y_train)
#model will predict based on given X_test
y_pred = model.predict(X_test)
#Plotting Confusion Matrix
class_names = ['Benign','Malignant']
plot_confusion_matrix(y_test,y_pred,class_names)
```

```python
#Super scary sounding RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier #1. Change which model we import
from sklearn.model_selection import train_test_split
#converts M,B to 1,0 respectively
y = np.array(df['diagnosis']=='M').astype(int)

X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,1:],df['diagnosis'])

#instantiating the model object
model = RandomForestClassifier(n_estimators=20) #2. Change which model we instantiate
#fitting/training model on train data
model.fit(X_train,y_train)
#model will predict based on given X_test
y_pred = model.predict(X_test)
#Plotting Confusion Matrix
class_names = ['Benign','Malignant']
plot_confusion_matrix(y_test,y_pred,class_names)

#We only had to change 2 lines from the previous logistic regression cell!
```

#### A note about using Sklearn and Modelling
The above cells demonstrated the power of abstraction - of hiding away the underlying mechanics in order to do incredibly powerful operations without needing to understand them completely. Abstraction is an important idea and I don't want to cut it short. But when diving into these kinds of predictive models, it is important to deeply understand the type of problem you are working with. In particular, Python has a robust library ecosystem, with most tech companies releasing data science packages in Python([Pyro](http://pyro.ai/) from Uber, [Pytorch](http://pytorch.org/) from Facebook, [Tensorflow](http://pytorch.org/) from Google). In short, Python is much simpler to understand, has incredible support systems built in for optimization and modelling, and is syntactically succinct(unlike R).

With that said, there are some  downfalls to Python that you should be aware of. Now, none of these should dissuade you from learning Python. Hopefully I've made my love for this language clear above. But as one dives deeper into Data Science or programming in general, these shortcomings will become more important and one should be aware of them.

1) Less intuitive Object Oriented Programming

This is not necessarily a bad thing. But OOP is not built into Python the way it is for, say java, where the language is designed around classes. Given the broad usage of this paradigm, it is important to understand and be experienced with Object orientation and sometimes Python allows you to slip into functional programming without ever touching classes.

2) Poor memory management

This one is a biggie. Python's dynamic typing and relatively higher level of abstraction means that it is much harder to maintain fine control over memory. This is also a result of the fact that Python is interpreted rather than compiled.

3) It's slow(relatively)

Speed tests reveal that Python can be orders of magnitude slower than other languages, like C/C++. But these orders of magnitude are relative - for the most part, the speed doesn't matter until operations become very computationally expensive or the datasets become too larger.

If you want to read more about the reasons for why Python's performance can be poorer compared to other languages, check out this [link](jakevdp.github.io/blog/2014/05/09/why-python-is-slow/)

Still interested in learning more about programming languages? Take [CS164](https://inst.eecs.berkeley.edu/~cs164/archives.html) at Berkeley!

#### Writing Pythonic (and just good) Code - a quick overview
1) Variables should be named in such a manner that their purpose and origin are easily identifiable

2) Functions and methods should have clear documentation using ''' triple quotes describing the input parameters, output and the functions purpose

3) Code should be written for other humans to read, not a computer
